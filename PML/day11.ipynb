# **paired t test**

This test is used when we have two paired sample the. average differance between paired value is compared against claim value to esatablish wheather the differance is statistically significant.

the keyword here is before-after

```python
before=np.array([12,14,10,9,8,13])
```

```python
after=np.array([11.8,9.7,11.3,8.2,7.2,12.3])
```

```python
#H0 avg(before-after)>=2
#Ha avg(before-after)<2  ====>>>left tail test
```

```python
d = (before - after)
```

```python
d
```

```python
np.mean(d)
```

```python
ssw.ttost_paired(before,after,low=2,upp=2)        #in ouput first 3 are right tail test and last 3 are left tail test. and 5 is defree if freedom
```

```python
scipy.stats.ttest_1samp(before - after,popmean=2, alternative='less')
```

# **4. One Way ANOVA.  ===> ANalysis Of VAariance**

This test is used when we have multiple sample. (more than 2)

This test will compared all the paired of the mean and check wheather they can be consider is equal or not

NULL hypothesis all the paired of means are equal

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
import statsmodels.stats.multicomp
```

```python
dfA = pd.DataFrame({'Col1':np.repeat('A',7),'Col2':np.random.randint(10,30,7)})
dfB = pd.DataFrame({'Col1':np.repeat('B',9),'Col2':np.random.randint(20,40,9)})
dfC = pd.DataFrame({'Col1':np.repeat('C',8),'Col2':np.random.randint(15,30,8)})
df = pd.concat([dfA,dfB,dfC])
```

```python
df
```

```python
mod1=ols('Col2 ~ Col1',df).fit()      #ols = ordinary list square
```

```python
tb1=sm.stats.anova_lm(mod1)
```

```python
tb1
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
print(pairwise_tukeyhsd(df.Col2, df.Col1))
```
