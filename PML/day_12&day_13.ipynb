**1-sample z test/ 1 sample t test**

This test is used to compare the sample mean of one sample sample against a calim value and establish whether their difference is statistically significant or not

**2-sample t test**

This test is used to compare the difference of means of two independent samples against a claim value and establish whether their difference is statistically significant or not

**Paired t test**

This test is used when we have two paired samples. The average difference between the paired values is compared against a claim values and establish whether their difference is statistically significant or not
The key word here is before-after

**One way ANOVA**

Used when we have multiple samples. This test compares all the pair of the means. It checks whether the difference of each pair of mean can be ignored or not OR can be considered to be equal to 0 or not.
The Ho in this case is that all the pair of means are equal

**Multi way ANOVA**

Used when we have multiple factors influencing the response. IN addition to their individual effect, we would also like to study the interactions between various factors.

# **Chi square test for goodness of fit**

Used when there are multiple categories. The counts obtained for each category would be compared aginst the claimed pattern to check whether the counts follow the pattern or not. The standard Ho is that the counts follow the pattern

```python
# claimed pattern for the qualifications
# Btech -0.4, Mtech -0.1, MCA -0.3, Msc -0.2
# Btech-50 , Mtech-20, MCA-45, MSC-25

# wheather counts obtained from sample follow the pattern or not
```

```python
count=np.array([50,20,45,25])
pattern = np.array([0.4,0.1,0.3,0.2])
```

```python
exp_counts = sum(count) * pattern
```

```python
exp_counts
```

```python
from scipy.stats import chisquare
```

```python
scipy.stats.chisquare(count,f_exp =exp_counts)
```

# **One sample possion rate**

This test is used when wish to compare the rate of which. event against the claim value and which establish wheather that differance is statistically significat or not

```python
# A good class would ask atleast 20 question every day
# H0 n>20
# Ha n<20   left tail test
```

```python
questions=[14,19,24,18,17,23,21,18,19,17]
```

```python
test_poisson(sum(questions),len(questions),20,method='score',alternative='smaller')
```

# **Two sample possion rate**

Used when we have two samples and the difference of the average counts of an event has to be compared against a claim value and establish whether the difference is statistically significant or not.

# **One Variance Test**

Used when the variance of a sample has to be compared against a claim value and establish whether their difference from 1 is statistically significant or not

**practical signoficant** >>>> changes from person to person




```python
df=pd.read_excel('CDAC_DataBook (1).xlsx',sheet_name='faithful')
```

```python
erupt=df.eruptions
```

```python
np.var(erupt,ddof=1)      # variance
```

```python
# H0 variance (erupt) < 1.1
# H0 variance (erupt) <= 1.1
# H0 variance (erupt) > 1.1   right tail test
# TS =(n-1)* variance / (claim value)
```

```python
ts = (len(erupt)-1)*np.var(erupt,ddof=1)/1.1          #test statistics
```

```python
ts
```

```python
from scipy.stats import chi2
```

```python
p_value = 1 - chi2.cdf(ts,len(erupt)-1).      # this formula is used for convert test statistics to p values
```

```python
p_value
```

```python
p_value =chi2.sf(ts,len(erupt)-1)
```

```python
p_value
```
