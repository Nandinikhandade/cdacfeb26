# ONE HOT ENCODING :

USED FOR NOMINAL SCALE VARIABLES (NO LOGICAL ORDER)

NOT USEFUL WHEN NUMBER OF CATEGORIES ARE TOO LARGE

REGRESSION ,ANN,KNN,SVM

# ORDINAL ENCODING

VARIABLES WITH INHERENT RANKING OR RATING

PRESERVE LOGICAL ORDER

```python
df = pd.DataFrame({
    'Education':[
        'Graduate',
        'PG',
        'High School',
        'Graduate',
        'PG'
    ]
})

education_map = {
    'High School':1,
    'Graduate':2,
    'PG':3
}

df['Education_Encoded'] = (
    df['Education']
    .map(education_map)
)

print(df)
```

# **Frequency Encoding**

```python
# Frequency encoding
df = pd.DataFrame({
    'City':[
        'Mumbai',
        'Mumbai',
        'Mumbai',
        'Delhi',
        'Delhi',
        'Pune'
    ]
})

freq = df.City.value_counts()

df['City_Frequency'] = (df.City.map(freq))
```
