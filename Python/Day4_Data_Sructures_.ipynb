```python
#Shopping list application
Insert an element, delete an element, print in list.
While inserting check if it does not exist; then only insert.
While deleting check if it exists then only delete.
Print is a normal printing listing all the elements entered in the shopping list.
```

```python
def menu():
    print("\n-----------------------")
    print("   My Shopping List App")
    print("-----------------------")
    print("1.Enter element in List")
    print("2.Delete element in List")
    print("3.Print The List")
    print("4.Exit")
    print("-----------------------")

def add_item(item,shopping_list):
    if item not in shopping_list:
        shopping_list.append(item)
        print(f"{item} added")
    else:
        print(f"{item} already there")
def print_list(shopping_list):
    print("\nItems in list are")
    print("-------------------")
    for index in range(len(shopping_list)):
        print(f"{index+1}.{shopping_list[index]}")

def delete_list(shopping_list):
    print_list(shopping_list)
    item_number=int(input("Enter item number to delete:"))
    if 0<item_number<=len(shopping_list):
        print(f"{shopping_list.pop(item_number-1)} deleted")
    else:
        print(f"{item_number} seems to be out of range.")
```

```python
shopping_list=[]
choice=0
while True:
    menu()#We will print all the options of the menu.
    choice=int(input(":"))
    if choice==1:
        item=input("Enter Item to add:")
        add_item(item,shopping_list)
    elif choice==2:
        delete_list(shopping_list)
    elif choice==3:
        print_list(shopping_list)
    elif choice==4:
        print("Thanks for using the system.")
        break
    else:
        print("Invalid option.")


```

```python
options={"tea":10.00,"coffee":20.00,"water":15.00,"breadtost":25.00}
order=[]

def menu():
    print("-------------------------")
    print("        APNA TAPRI       ")
    print("-------------------------")
    i=1
    for key,value in options.items():
        print(f"{i}.{key}....{value}")
        i+=1
    print("-------------------------")
def billing(order):
    sub_total=0.0
    print("\n-------------------")
    print("          BILL      ")
    print("\n-------------------")
    for item,quantity in order.items():
        print(f"{item} X {quantity} = {quantity*options[item]}")
        sub_total+=quantity*options[item]
    print("\n-------------------")
    print(f"  Sub Total:{sub_total}")
    if sub_total>=100:
        disscount=sub_total*0.10
    else:
        disscount=0.0
    print("\n-------------------")
    print(f"  Disscount:{disscount}")
    print(f"  Total    :{sub_total-disscount}")
def place_order():
    order={}
    while True:#User does not type exit.
        item=input("Enter item(EXIT to stop):")
        if item=="EXIT":
            break
        qty=int(input("Enter quantity:"))
        if item in order:
            #Check if the item already exists in the order. If so update the quantity.
            order[item]=order[item]+qty
        else:
            order[item]=qty
    billing(order)
    #call billing pass order:order






    #Check item order, take their prices from options
    #Take items from order and check their quantity from the order given.
    #Generate a bill for the items ordered, also the 10% discount if the bill is above ₹100.



while True:
    menu()
    choice=int(input("Enter 1.Order 2.Exit"))
    if choice==1:
        place_order()
    elif choice==2:
        break
    else:
        print("Please check the option.")



```

```python
'''
#From a given line of string, count:
- number of digits
- number of upper case
- number of lower case
- number of symbols
- number of words
'''
line=input("Enter line:")
digit_counter=upper_counter=lower_counter=symbol_counter=word_counter=0
for char in line:
    if char.isupper():
        upper_counter+=1
    elif char.islower():
        lower_counter+=1
    elif char.isdigit():
        digit_counter+=1
    elif not char.isalpha() and not char.isdigit() and not char.isspace():
        symbol_counter+=1
    elif char.isspace():
        word_counter+=1
print(f"digits:{digit_counter}\nUpperCase:{upper_counter}\nLowerCase{lower_counter}\nSymbols:{symbol_counter}\nWords:{word_counter+1}")
```
