```python
#Custom or User-Defined Exception
#Generally designed for specific project-based condition which will occur. Example: for voting system under 18, may cause exception.

to design:
1. Create class with in herit exception class.
2.Class created in step 1 would have two methods:
    1. __init__(), which will take the message of exception
    2. __str__(), which will print the message
3. Since they are custom design exceptions, the user has to raise it on their own. done by 'raise'
```

```python
class AgeException(Exception):
    def __init__(self,age):
        self.age=age
    def __str__(self):
        return f"AgeException raised: at {self.age} under age detected"
```

```python
try:
    print("Welcome to Voting System.")
    age=int(input("Enter your age:"))
    if age<18:
        raise AgeException(age)
    else:
        print("Welcome to vote:")
except Exception as msg:
    print("Error:",msg)
```

```python
Create custom exception where the user can raise:
ProjectException
- 404 error for file not found
- 500 error as user not found
- unknown error if the number is not known
```

```python
class ProjectException(Exception):
    def __init__(self, error_code):
        self.error_code = error_code

        if error_code == 404:
            self.message = "404 Error: File Not Found"
        elif error_code == 500:
            self.message = "500 Error: User Not Found"
        else:
            self.message = "Unknown Error: Error code not recognized"

    def __str__(self):
        return self.message
try:
    code=input("Enter code")
    raise ProjectException(code)
except Exception as e:
    print(e)
```
