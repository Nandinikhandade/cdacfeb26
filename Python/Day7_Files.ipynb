```python
from os import X_OK
File:
Majorly used for permanent storage or passing data from one module to another via file
Large-scale data can be stored or processed via a file.
process:
    1.open:
        <ref>=open("file/path","mode")
        fw=open("my_file.txt","w")
        r:read only(should already exist; otherwise fails.)
        w:write only(if the file does not exist, it will create the file but if the file is existing, it will overwrite and create a new file.)
        a: append only at the end of the file. (Popular because if the file does not exist it will create it; if it exists it will add data at the end.)
        r+:read and write(you can also write or overwrite the content of a file.writing default done from start of file. Also capable of reading the start-to-end content of a file.)
        w+:write and read(write your read will write either to a new file or a newly created version of a file but whatever was recently written by the write command can be read again.)
        a+:append and read(writes to the end of the file but is capable of reading from start to end of the file.)
        x: it will open a file in a write mode. If it is not available, it will create but if it is already available, it will generate an error.

2. Depending on the need, operations like reading, writing would be done.
    to read:
            read(): single shot: read the entire content of a file.
            read(buffer): at a single shot willl only read buffered data. At a time it will only read data of the given buffer size.
            readline(): read one line at a time.
            readlines(n): can read n given lines in a single read.
    to write:
            write(data)#Needed data to be converted to string most of the time for formatted writing.
3.close:
        .close(): only when called.close, the changes made to the file by reading, writing, or appending will be made permanent and saved.
```

```python
#Content of a file line by line and stopping automatically
try:
    file=input("Name of the file")
    fr=open(file,"r")
    while True:
        line=fr.readline()
        if line=="":#Line will return blank when EOF is reached.
            print("EOF".center(80,"-"))
            break
        print(line)
except FileNotFoundError:
    print("File not found")
else:
    fr.close()
```

```python
#Write a code to accept two source filenames from the user and create a new file called combine.txt, which will contain the contents of both the files in an alternate manner.
#line 1
#a line
#line 2
file1 = input("Enter first filename: ")
file2 = input("Enter second filename: ")

try:
    f1 = open(file1, "r")
    f2 = open(file2, "r")
    fw = open("combine.txt", "a")

    while True:
        line1 = f1.readline()
        line2 = f2.readline()

        if line1 == "" and line2 == "":
            break

        if line1 != "":
            fw.write(line1)

        if line2 != "":
            fw.write(line2)

    print("Files combined successfully into combine.txt")

    f1.close()
    f2.close()
    fw.close()

except FileNotFoundError:
    print("One of the files was not found")
```

```python
Create a function which will accept a file name and a secret key. First it will encrypt the file and give it back. If an encrypted file is given, it will give you the decrypted file.
```

```python
import getpass

def process_file(filename, key):

    try:
        fr = open(filename, "r")
        data = fr.read()
        fr.close()

        result = ""

        for char in data:
            result += chr(ord(char) ^ key)

        new_file = "processed_" + filename

        fw = open(new_file, "w")
        fw.write(result)
        fw.close()

        print("Processing completed.")
        print("Output file:", new_file)

    except FileNotFoundError:
        print("File not found.")


# main program
file = input("Enter file name: ")
key = int(getpass.getpass("Enter secret key: "))

process_file(file, key)
```
