```python
Create a class Time. Have a method setTime which sets HH:MM:SS
Also has a method display which will display time in hours minutes and seconds.
Create a class with methods:
- increase_hour
- increase_minute
- increase_second
which will respectively increase the given unit.
```

```python
class Time:
    def set_time(self,h,m,s):
        self.h=h
        self.m=m
        self.s=s
        print(f"Time set to:{self.h}:{self.m}:{self.s}")
    def increase_hour(self):
        self.h+=1
        print(f"New time set to:{self.h}:{self.m}:{self.s}")
    def increase_minute(self):
        self.m+=1
        if self.m>=60:
            self.h+=1
            self.m=self.m%60
        print(f"New time set to:{self.h}:{self.m}:{self.s}")
    def increase_second(self):
        self.s+=1
        if self.s>=60:
            self.m+=1
            self.s=self.m%60
        print(f"New time set to:{self.h}:{self.m}:{self.s}")
```

```python

class Account:
    account_number_generator=202600
    def __init__(self,name,amount):
        self.name=name
        self.amount=amount
        Account.account_number_generator+=1
        self.account_number=Account.account_number_generator
        print(f"Account successfully created. Your account number is {self.account_number}")
    def check_balance(self):
        print(f"your account_number:{self.account_number} has {self.amount}. ")
    def deposite(self,amount):
        if amount>0:
            self.amount+=amount
            self.check_balance()
        else:
            print(f"{amount} not be negative.")
    def withdraw(self,amount):
        if self.amount-amount>3000:
            self.amount-=amount
            self.check_balance()
        else:
            print(f"Low balance, hence transaction cannot be performed")
    def get_account(self):
        return self.account_number
```

```python
bank={}
class Account:
    account_number_generator=202600
    def __init__(self,name,amount):
        self.name=name
        self.amount=amount
        Account.account_number_generator+=1
        self.account_number=Account.account_number_generator
        print(f"Account successfully created. Your account number is {self.account_number}")
    def check_balance(self):
        print(f"your account_number:{self.account_number} has {self.amount}. ")
    def deposite(self,amount):
        if amount>0:
            self.amount+=amount
            self.check_balance()
        else:
            print(f"{amount} not be negative.")
    def withdraw(self,amount):
        if self.amount-amount>3000:
            self.amount-=amount
            self.check_balance()
        else:
            print(f"Low balance, hence transaction cannot be performed")
    def get_account(self):
        return self.account_number

def menu():
    print(f'''\n{"".center(30,'-')}''')
    print("LENA BANK".center(30))
    print(f'''{"".center(30,'-')}''')
    print("1.Create Account")
    print("2.Deposite")
    print("3.Withdraw")
    print("4.Check Balance")
    #Fifth option: Account Transfer, where the user has to give the source account, destination account, and transfer money from the source account to the destination account, checking the minimum balance option.
    print("0.Exit")
    print(f'''{"".center(30,'-')}''')

choice=0
while True:
    menu()
    choice=int(input(":"))
    if choice==1:
        #An account only if 5,000 or 5,000+ is given.
        name=input("Enter Name:")
        while True:
            amount=float(input("Enter Amount"))
            if amount>=5000:
                a=Account(name,amount)
                break
            else:
                print("To open an account one needs a minimum of 5,000.")
        bank[a.get_account()]=a
    elif choice==2:
        account=int(input("Enter account number:"))
        if account in bank:
            print("{account} found...")
            bank[account].deposite(float(input("Enter amount:")))
        else:
            print("{account} not found...")
    elif choice==3:
        account=int(input("Enter account number:"))
        if account in bank:
            print("{account} found...")
            bank[account].withdraw(float(input("Enter amount:")))
        else:
            print("{account} not found...")
    elif choice==4:
        account=int(input("Enter account number:"))
        if account in bank:
            print("{account} found...")
            bank[account].check_balance()
        else:
            print("{account} not found...")
    elif choice==0:
        print("Thank you for banking with us.")
        break
    else:
        print("Selected option is invalid.")
```

```python
class A:
    def __init__(self):
        print("Initializing A")

class B(A):
    def __init__(self):
        print("Initializing B")
        super().__init__()

class C(A):
    def __init__(self):
        print("Initializing C")
        super().__init__()

class D(B, C):
    def __init__(self):
        print("Initializing D")
        super().__init__()

# Instantiate D
d = D()
#It is thanks to MRO method resolution order.
#These in it are called:
'''
1. When super is initiated, the first super class B is called.
2. The second super class C is called.
3. From B and C, their super class A is called.'''
```

```python
#create human:has gender ,name
# education : has degree
# job: has salary
#using __init__() set all values and print them from job only
#accept parameters in job only but set in respective class
```

```python
class Human:
    def __init__(self, name, gender):
        self.name = name
        self.gender = gender
        print("name,gender set in Human")

class Education(Human):
    def __init__(self, name,gender,degree):
        super().__init__(name,gender)
        self.degree = degree
        print("degree set in Education")

class Job(Education):
    def __init__(self, name, gender, degree, salary):
        super().__init__(name, gender,degree)
        self.salary = salary
        print("salary set in Job ")

    def display(self):
        print(f"Name: {self.name}")
        print(f"Gender: {self.gender}")
        print(f"Degree: {self.degree}")
        print(f"Salary: {self.salary}")

employee = Job("Amit", "Male", "B.Tech", 60000)
employee.display()
```
