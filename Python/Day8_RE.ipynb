```python
import re
p=r"\d{10}"
word=input("Enter:")
if re.search(p,word):
    print("Accepted")
else:
    print("Rejected")
```

```python
#End with digit.
import re
p=r"\d$"
word=input("Enter:")
if re.search(p,word):
    print("Accepted")
else:
    print("Rejected")
```

```python
#End with digit.
import re
p=r"^\d"
word=input("Enter:")
if re.search(p,word):
    print("Accepted")
else:
    print("Rejected")
```

```python
#Accept date of birth precisely: DD-MM-YYYY.
#Start with two digits, then a hyphen, then two digits, then a hyphen, and we end with four digits.
#End with digit.
import re
p=r"^\d{2}-\d{2}-\d{4}$"
word=input("Enter:")
if re.search(p,word):
    print("Accepted")
else:
    print("Rejected")
```

```python
#Start with A to Z, capital or lower, but must end with a digit.
import re
p=r"^[A-Za-z].*\d$"
word=input("Enter:")
if re.search(p,word):
    print("Accepted")
else:
    print("Rejected")
```

```python
#A username should be at least 5 characters and at most 12 characters and is only allowed to use A to Z (upper or lower) and digits
import re
p=r"^\w{5,12}$"
word=input("Enter:")
if re.search(p,word):
    print("Accepted")
else:
    print("Rejected")
```

```python
#Regular expression to validate email.
import re
p=r"^\w+@\w+\.\w+$"
word=input("Enter:")
if re.search(p,word):
    print("Accepted")
else:
    print("Rejected")
```
