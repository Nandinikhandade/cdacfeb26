# **There are three forms of statistics:**

**1.Descriptive Statistics**

Used to summarize and describe data.

**Examples:** mean, median, mode, charts, graphs, standard deviation.

**2.Inferential Statistics**

Used to make conclusions or predictions about a population based on sample data.

**Examples:** hypothesis testing, confidence intervals, regression analysis.

**3.Predictive Statistics**

Used to predict future outcomes using historical data and statistical models.

**Examples:** machine learning models, forecasting, predictive analytics.

```python
n=3000
xbar=1800
stdev=70
cl=0.95

norm.ppf((1-cl)/2)

z=np.abs(norm.ppf((1-cl)/2))    # critical value of x
z

t=np.abs(t.ppf((1-cl)/2,n-1))  # t distribution requires degree of freedomm thats why n-1
t

low_est =xbar - t*stdev/n**0.5
upper_est=xbar + t*stdev/n**0.5

size =upper_est - low_est
size
```

```python
stats.t.interval(cl,n-1,xbar,stdev/n**0.5)             # output(lower estimate, upper estimate)
```
